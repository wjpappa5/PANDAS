```python
# Import Dependencies
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reference the file where the CSV is located
file = "/Users/williampappas/Desktop/GitLab/01-Class-Activities/04-Python-Pandas/2/Activities/04-Par_PortlandCrime-Cleaning/Unsolved/Resources/crime_incident_data2017.csv"
# Import the data into a Pandas DataFrame
original_df = pd.read_csv(file)
original_df.head()
```

```python
# look for missing values
original_df.count()
```

```python
# drop null rows
df = original_df.dropna(how='any')
```

```python
# verify counts
df.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
df['Offense Type'].value_counts()
```

```python
# Combine similar offenses
df['Offense Type'] = df['Offense Type'].replace(
    {'Theft From Motor Vehicle': 'Vehicle Type Theft', 'Motor Vehicle Theft': 'Vehicle Type Theft', 'Theft of Motor Vehicle Parts or Accessories':'Vehicle Type Theft'})
```

```python
# Check to see if you comnbined similar offenses correctly in "Offense Type
df['Offense Type'].value_counts()
```

```python
# Create a new DataFrame that looks into a specific neighborhood
grouped_crime_df = df.set_index("Neighborhood")
grouped_crime_df.head()
```
