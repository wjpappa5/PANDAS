```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merged_df = pd.merge(bitcoin_df, dash_df, on="Date")
merged_df.head()
```

```python
# Rename columns so that they are differentiated
merged_df = merged_df.rename(
    columns={"Open_x": "Bitcoin Open", "High_x":"Bitcoin Hight","Low_x":"Bitcoin Low","Close_x":"Bitcoin Close","Volume_x":"Bitcoin Volume","Market Cap_x":"Bitcoin Market Cap","Open_y": "Dash Open", "High_y":"Dash Hight","Low_y":"Dash Low","Close_y":"Dash Close","Volume_y":"Dash Volume","Market Cap_y":"Dash Market Cap"})
merged_df.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
```

```python
# Collecting best open for Bitcoin and Dash
bestbitcoin= merged_df["Bitcoin Open"].max()
bestdash= merged_df["Dash Open"].max()
# Collecting best close for Bitcoin and Dash
closebitcoin= merged_df["Bitcoin Close"].max()
closedash= merged_df["Dash Close"].max()
# Collecting the total volume for Bitcoin and Dash
totalbitcoin= merged_df["Bitcoin Volume"].sum()
totaldash= merged_df["Dash Volume"].sum()
```

```python
# Creating a summary DataFrame using above values
crypto_table = pd.DataFrame({"Best Bitcoin Open": [bestbitcoin],
                              "Best Dash Open": [bestdash],
                              "Best Bitcoin Close": [closebitcoin],
                              "Best Dash Close": [closedash],
                               "Total Bitcoin Volume": [totalbitcoin],
                              "Total Dash Volume" : [totaldash],
                              })
crypto_table = crypto_table[["Best Bitcoin Open","Best Dash Open","Best Bitcoin Close","Best Dash Close","Total Bitcoin Volume","Total Dash Volume"]]
crypto_table["Total Bitcoin Volume"] = crypto_table["Total Bitcoin Volume"].map('${:,.2f}MM'.format)
crypto_table["Total Dash Volume"] = crypto_table["Total Dash Volume"].map('${:,.2f}MM'.format)
crypto_table
```
