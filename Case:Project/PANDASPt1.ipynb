```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "/Users/williampappas/Desktop/GitLab/02-Homework/04-Python-Pandas/03-Pandas-Project/Mini-Project-Part-1/Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
coder_data = pd.read_csv(file, encoding= 'iso-8859-1',low_memory=False)
coder_data.head()
```

```python
# Inspect all columns
coder_data.count()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
adjusted_coder = coder_data.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
adjusted_coder.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
adjusted_coder = adjusted_coder.replace({0.0: "No", 1.0: "Yes"})
adjusted_coder.head()
```

```python
# Calculate total number of respondents in survey
total_respondents = len(adjusted_coder)
total_respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attendance = adjusted_coder.loc[adjusted_coder["AttendedBootcamp"] == "Yes"]
attendance.count()
```

```python
# Calculate average age of attendees
average_age = attendance["Age"].mean()
# Calculate how many people attended a bootcamp
numattend = attendance["AttendedBootcamp"].count()
# Calculate how many attendees hold degrees
degrees = attendance["SchoolDegree"].count()
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender = attendance["Gender"].count()
male_gender = attendance["Gender"].value_counts()['male']
female_gender = attendance["Gender"].value_counts()['female']
non_gender = total_gender - male_gender - female_gender
# Calculate percentage of respondents who attended a bootcamp
percattend = round((numattend / total_respondents) * 100,2)
# Calculate percentage of respondents belonging to each gender
percmale = round((male_gender / total_gender) * 100,2)
percfemale = round((female_gender / total_gender) * 100,2)
percnon = round((non_gender / total_gender) * 100,2)
# Calculate percentage of attendees with a school degree
percdegree = round((degrees / numattend) * 100,2)
# Calculate average post-bootcamp salary
avgsalary = attendance["BootcampPostSalary"].mean()
```

```python
# Create a new table consolodating above calculations
bootcamp_table = pd.DataFrame({"Total Surveyed": [total_respondents],
                              "Total Attendees": [numattend],
                              "Average Age Attended": [average_age],
                              "Attendance with Degrees": [degrees],
                               "Percent with Degree": [percdegree],
                              "Males Attended" : [male_gender],
                              "Male Percent": [percmale],
                              "Females Attended": [female_gender],
                              "Female Percent": [percfemale],
                              "Non Gender Attended": [non_gender],
                              "Non Gender Percent": [percnon],
                               "Average Post Salary": [avgsalary]
                              })
bootcamp_table = bootcamp_table[["Total Surveyed","Total Attendees","Average Age Attended","Attendance with Degrees","Percent with Degree","Males Attended","Male Percent","Females Attended","Female Percent","Non Gender Attended","Non Gender Percent","Average Post Salary"]]
bootcamp_table = bootcamp_table.round(2)

bootcamp_table
```

```python
# Improve formatting before outputting spreadsheet
bootcamp_table["Percent with Degree"]=bootcamp_table["Percent with Degree"].map("{0:,.2f}%".format)
bootcamp_table["Male Percent"]=bootcamp_table["Male Percent"].map("{0:,.2f}%".format)
bootcamp_table["Female Percent"]=bootcamp_table["Female Percent"].map("{0:,.2f}%".format)
bootcamp_table["Non Gender Percent"]=bootcamp_table["Non Gender Percent"].map("{0:,.2f}%".format)
bootcamp_table["Average Post Salary"]=bootcamp_table["Average Post Salary"].map("{0:,.2f}$".format)
bootcamp_table
```
